```python
import csv
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot  as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D    #Para lo graficos 3d
```

Defino las funciones `separar()` y `separarletra()` empleadas para separar los valores de la variables aleatoria

```python
def separarletra(val):
    entero=0
    for i in range(1,len(val)):
        entero=entero*10
        entero+=int(val[i])        
    return entero

def separar(lista):
    vector=[]
    for etr in lista:
        if len(etr)>0:
            vector.append(separarletra(etr))
        else:
            continue
    return vector
```

Extracción de los datos y el cálculo de las sumas de columnas y filas para la obtencion de la funcion de densidad marginal de ambas variables aleatorias, esto empleando la librería `numpy`, de los datos se obtiene la lista que contiene los datos de probabilidad expuestos en una matriz, además de los valores que toman las variables **X** y **Y**.

```python
datos=[]
valoresX=[]
valoresY=[]
with open('xy.csv', newline='') as File:  #Abrimos el documento
    lineas = csv.reader(File)               #Extraemos cada linea en una lista
    for row in lineas:
        datos.append(row)
```

```python
datos_mod=np.array(datos)

#Extraccion de los datos de X y Y
valoresX=separar(datos_mod[:,0])
valoresY=separar(datos_mod[0,:])

#Filtrado de datos
datos_mod=np.delete(datos_mod, [0][0], axis=1)
datos_mod=np.delete(datos_mod, [0][0], axis=0)

datos_mod=np.array(datos_mod, dtype=np.float64)  #Matriz de datos de tipo flotante

sumaY=np.sum(datos_mod, axis=0)   #Valores de la funcion de densidad marginal de Y
sumaX=np.sum(datos_mod, axis=1)   #Valores de la funcion de densidad marginal de X

print("La suma para el eje x para la funcion de densidad de X es: ", np.sum(sumaX))
print("La suma para el eje y para la funcion de densidad de Y es: ", np.sum(sumaY))
```

Gráfica de la función de densidad marginal para **X**

```python

plt.plot(valoresX, sumaX)
plt.xlabel("Valores de X")
plt.ylabel("Probabilidad")
plt.title("Funcion de densidad marginal para X")
plt.savefig("Graficas/Grafica_para_X.png")
```

Gráfica de la función de densidad marginal para **Y**

```python
plt.plot(valoresY, sumaY)
plt.xlabel("Valores de Y")
plt.ylabel("Probabilidad")
plt.title("Funcion de densidad marginal para Y")
plt.savefig("Graficas/Grafica_para_Y.png")
```

### Cálculo de las curvas de mejor ajuste para la funcion de densidad marginal de cada variable
Para este paso se evaluan las cuatro distribuciones más comunes las cuales son la ditribución *Normal*, *Rayleigh*, *Exponencial* y la *uniforme*, por lo tanto para lograr este objetivo se creó una función de python para cada distribución, además se hace uso de la funcion `curve_fit()` de la librería `scipy.optimize`, la cual toma como parametros la funcion de la ditribucion y los valores de *x* y *y*, de la funcion con la que se quiere ajustar y devuelve los parametros de la distribucion que mejor se justan a los valores de la funcion con la que se esta comparando.

```python
def gaussiana(x, a, b):
    return (1/(np.sqrt(2*(np.pi*(b*b)))))*np.exp(-(((x-a)*(x-a))/(2*(b*b))))


#Parametros para X
paramX,_=curve_fit(gaussiana,valoresX,sumaX)
print("Los parametros del ajuste de distribucion normal para x son: \n",paramX)

#Parametros para Y
paramY,_=curve_fit(gaussiana,valoresY,sumaY)
print("Los parametros del ajuste de distribucion normal para y son: \n", paramY)

#Genero los valores para los plots para cada funcion de densidad marginal

Xys=gaussiana(valoresX,paramX[0],paramX[1])

Yys=gaussiana(valoresY,paramY[0],paramY[1])
```

Gráfica de la distribución normal para la función de densidad marginal de **X**:

```python
plt.plot(valoresX, sumaX)
plt.plot(valoresX,Xys)
plt.xlabel("Valores de X")
plt.ylabel("Probabilidad")
plt.title("Curva de ajuste gaussiana para X")
plt.savefig("Graficas/Grafica_para_X_Curva_de_ajuste_normal.png")
```

Gráfica de la distribución normal para la función de densidad marginal de **Y**:

```python
plt.plot(valoresY, sumaY)
plt.plot(valoresY,Yys)
plt.xlabel("Valores de Y")
plt.ylabel("Probabilidad")
plt.title("Curva de ajuste gaussiana para X")
plt.savefig("Graficas/Grafica_para_Y_Curva_de_ajuste_normal.png")
```

Ahora calculamos para la distribución de tipo *Rayleigh*

```python
def rayleigh(x, s):
    return (x/(2*pow(s,2)))*np.exp((-(pow(x,2)))/(2*pow(s,2)))

#Parametros para X
paramX_rayleigh,_=curve_fit(rayleigh,valoresX,sumaX)
print("Los parametros del ajuste de distribucion rayleigh para x son: ", paramX_rayleigh)

#Parametros para Y
paramY_rayleigh,_=curve_fit(rayleigh,valoresY,sumaY)
print("Los parametros del ajuste de distribucion rayleigh para y son: ", paramY_rayleigh)

#Genero los valores para los plots para cada funcion de densidad marginal

Xys_rayleigh=[]
Yys_rayleigh=[]

for i in range(0,len(valoresX)):
    Xys_rayleigh.append(rayleigh(valoresX[i], paramX_rayleigh[0]))    

for i in range(0,len(valoresY)):
    Yys_rayleigh.append(rayleigh(valoresY[i], paramY_rayleigh[0]))
```

Gráfica de la distribución Rayleigh para la función de densidad marginal de **Y**:

```python
plt.plot(valoresX, sumaX)
plt.plot(valoresX,Xys_rayleigh)
plt.xlabel("Valores de X")
plt.ylabel("Probabilidad")
plt.title("Curva de ajuste rayleigh para X")
plt.savefig("Graficas/Grafica_para_X_Curva_de_ajuste_rayleigh.png")
```

Gráfica de la distribución Rayleigh para la función de densidad marginal de **Y**:

```python
plt.plot(valoresY, sumaY)
plt.plot(valoresY,Yys_rayleigh)
plt.xlabel("Valores de Y")
plt.ylabel("Probabilidad")
plt.title("Curva de ajuste rayleigh para Y")
plt.savefig("Graficas/Grafica_para_Y_Curva_de_ajuste_rayleigh.png")
```

Ahora calculamos para la distribución de tipo *Exponencial*

```python
def exponencial(x,a):
    return a*np.exp(-a*x)

#Parametros para X
paramX_exp,_=curve_fit(exponencial,valoresX,sumaX)
print("Los parametros del ajuste de distribucion exponancial para x son: ", paramX_exp)

#Parametros para Y
paramY_exp,_=curve_fit(exponencial,valoresY,sumaY)
print("Los parametros del ajuste de distribucion exponancial para y son: ", paramY_exp)

#Genero los valores para los plots para cada funcion de densidad marginal

Xys_exp=[]
Yys_exp=[]

for i in range(0,len(valoresX)):
    Xys_exp.append(exponencial(valoresX[i], paramX_exp[0]))    

for i in range(0,len(valoresY)):
    Yys_exp.append(exponencial(valoresY[i], paramY_exp[0]))
```

Gráfica de la distribución Exponencial para la función de densidad marginal de **X**:

```python
plt.plot(valoresX, sumaX)
plt.plot(valoresX,Xys_exp)
plt.xlabel("Valores de X")
plt.ylabel("Probabilidad")
plt.title("Curva de ajuste exponencial para X")
plt.savefig("Graficas/Grafica_para_X_Curva_de_ajuste_exp.png")
```

Gráfica de la distribución Exponencial para la función de densidad marginal de **Y**:

```python
plt.plot(valoresY, sumaY)
plt.plot(valoresY,Yys_exp)
plt.xlabel("Valores de Y")
plt.ylabel("Probabilidad")
plt.title("Curva de ajuste exponencial para Y")
plt.savefig("Graficas/Grafica_para_Y_Curva_de_ajuste_exp.png")
```

Ahora calculamos para la distribución de tipo *Uniforme*

```python
def uniforme(x,a):
    return 1/a

#Parametros para X
paramX_uni,_=curve_fit(uniforme,valoresX,sumaX)
print("Los parametros del ajuste de distribucion uniforme para x son: ", paramX_uni)

#Parametros para Y
paramY_uni,_=curve_fit(uniforme,valoresY,sumaY)
print("Los parametros del ajuste de distribucion uniforme para y son: ", paramY_uni)

#Genero los valores para los plots para cada funcion de densidad marginal

Xys_uni=[]
Yys_uni=[]

for i in range(0,len(valoresX)):
    Xys_uni.append(uniforme(valoresX[i], paramX_uni[0]))    

for i in range(0,len(valoresY)):
    Yys_uni.append(uniforme(valoresY[i], paramY_uni[0]))
```

Gráfica de la distribución Uniforme para la función de densidad marginal de **X**:

```python
plt.plot(valoresX, sumaX)
plt.plot(valoresX,Xys_uni)
plt.xlabel("Valores de X")
plt.ylabel("Probabilidad")
plt.title("Curva de ajuste uniforme para X")
plt.savefig("Graficas/Grafica_para_X_Curva_de_ajuste_uni.png")
```

Gráfica de la distribución Uniforme para la función de densidad marginal de **Y**:

```python
plt.plot(valoresY, sumaY)
plt.plot(valoresY,Yys_uni)
plt.xlabel("Valores de Y")
plt.ylabel("Probabilidad")
plt.title("Curva de ajuste uniforme para Y")
plt.savefig("Graficas/Grafica_para_Y_Curva_de_ajuste_uni.png")
```

### Cálculo de la Covarianza, Correlacción y coeficiente de Correlación  a partir de los datos

Primero se comienza por el calculo de la media de cada variable, entonces aplicamos la ecuación:

<a href="https://www.codecogs.com/eqnedit.php?latex=\mu_{x}=\sum_{x}&space;xp(x)" target="_blank"><img src="https://latex.codecogs.com/gif.latex?\mu_{x}=\sum_{x}&space;xp(x)" title="\mu_{x}=\sum_{x} xp(x)" /></a>

Y para la variable **Y** aplicamos la misma ecuación:

<a href="https://www.codecogs.com/eqnedit.php?latex=\mu_{y}=\sum_{y}&space;yp(y)" target="_blank"><img src="https://latex.codecogs.com/gif.latex?\mu_{y}=\sum_{y}&space;yp(y)" title="\mu_{y}=\sum_{y} yp(y)" /></a>

Entonces dado que la *_p(x)_* y *_p(y)_* equivale a los valores que toma la función de densidad marginal de **X** y **Y** respectivamente se procede a calcular la media de **X** y **Y** como:


```python
media_x=0
media_y=0

for i in range(0,len(valoresX)):
    media_x+=valoresX[i]*sumaX[i]

for i in range(0,len(valoresY)):
    media_y+=valoresY[i]*sumaY[i]

print("La media de X es: ", media_x)
print("La media de Y es: ", media_y)
```

Ahora calculamos la varianza de las variables aleatorias empleando las ecuaciones:

<a href="https://www.codecogs.com/eqnedit.php?latex=V[x]&space;=&space;E[x^{2}]-\mu_{x}^{2}" target="_blank"><img src="https://latex.codecogs.com/gif.latex?V[x]&space;=&space;E[x^{2}]-\mu_{x}^{2}" title="V[x] = E[x^{2}]-\mu_{x}^{2}" /></a>

<a href="https://www.codecogs.com/eqnedit.php?latex=V[y]&space;=&space;E[y^{2}]-\mu_{y}^{2}" target="_blank"><img src="https://latex.codecogs.com/gif.latex?V[y]&space;=&space;E[y^{2}]-\mu_{y}^{2}" title="V[y] = E[y^{2}]-\mu_{y}^{2}" /></a>

Donde debido a que ya tenemos la media podemos calcularla como:

```python
varianza_x=0
varianza_y=0

for i in range(0,len(valoresX)):
    varianza_x=((valoresX[i])**2)*sumaX[i] - media_x

for i in range(0,len(valoresY)):
    varianza_y=((valoresY[i])**2)*sumaY[i] - media_y

print("La varianza de x es: ", varianza_x)
print("La varianza de y es: ", varianza_y)
```

Ahora calculamos la desviación de las variable **X** y **Y**, entonces obtenemos: 

```python
desviacion_x=np.sqrt(varianza_x)
desviacion_y=np.sqrt(varianza_y)

print("La desviacion de la variable X es: ", desviacion_x)
print("La desviacion de la variable y es: ", desviacion_y)
```

Ahora calcualmos el valor de la correlacion para las variables **X** y **Y**, esto empleando la sguiente ecuacion:

<a href="https://www.codecogs.com/eqnedit.php?latex=R_{xy}=E[XY]=\sum_{x}\sum_{y}xyp(x,y)" target="_blank"><img src="https://latex.codecogs.com/gif.latex?R_{xy}=E[XY]=\sum_{x}\sum_{y}xyp(x,y)" title="R_{xy}=E[XY]=\sum_{x}\sum_{y}xyp(x,y)" /></a>

Entonces obtenemos:

```python
correlacion=0
for i in range(0,len(valoresX)):
    for j in range(0,len(valoresY)):
        correlacion+=valoresX[i]*valoresY[j]*datos_mod[i][j]
print("La correlacion de las variables X y Y es: ", correlacion)
```

Entonces ahora podemos calcular la *Covarianza* de las variables **X** y **Y**, Yesto empleando la ecuación:
<a href="https://www.codecogs.com/eqnedit.php?latex=C_{xy}=E[XY]-\mu_{x}&space;\mu_{y}&space;=\sum_{x}\sum_{y}xyp(x,y)&space;-&space;\mu_{x}&space;\mu_{y}" target="_blank"><img src="https://latex.codecogs.com/gif.latex?C_{xy}=E[XY]-\mu_{x}&space;\mu_{y}&space;=\sum_{x}\sum_{y}xyp(x,y)&space;-&space;\mu_{x}&space;\mu_{y}" title="C_{xy}=E[XY]-\mu_{x} \mu_{y} =\sum_{x}\sum_{y}xyp(x,y) - \mu_{x} \mu_{y}" /></a>

La cual es básicamente la correlación menos la multiplicación de la medias de X y Y. Entonces obtenemos lo siguiente:

```python
covarianza=correlacion-(media_x*media_y)
print("La covarianza para las variables X y Y es: ",covarianza)
```

Para el cálculo del *coeficiente de correlacion (Pearson)*, se aplica la siguiente ecuación:

<a href="https://www.codecogs.com/eqnedit.php?latex=\rho&space;=\frac{C_{xy}}{\sigma_{x}&space;\sigma_{y}}" target="_blank"><img src="https://latex.codecogs.com/gif.latex?\rho&space;=\frac{C_{xy}}{\sigma_{x}&space;\sigma_{y}}" title="\rho =\frac{C_{xy}}{\sigma_{x} \sigma_{y}}" /></a>

Entonces obtenemos lo siguiente:

```python
pearson=(covarianza)/(desviacion_x*desviacion_y)
print("El coeficiente de correlacion (Pearson) es: ", pearson)
```

### Graficas en 3D
Ahora calculamos las gráficas en 3d de la función de densidad conjunta, obtiene la gráfica a partir de los datos y la gráfica a partir del modelo. Primero se obtiene los datos en formato de vectorial para ello empleamos el archivo `xyp.csv` el cual contiene los mismos datos que el archivo `xy.csv`, pero en un formato de 3 columnas lo cual facilita la obtención de los datos de probabilidad necesarios para realizar el gráfico 

```python
datos_3col=[]
with open('xyp.csv', newline='') as File:  #Abrimos el documento
    lineas = csv.reader(File)               #Extraemos cada linea en una lista
    next(lineas)
    for row in lineas:
        datos_3col.append(row)
        
x=[int(datos_3col[x][0]) for x in range(0,len(datos_3col)) ]
y=[int(datos_3col[x][1]) for x in range(0,len(datos_3col)) ]
z=[float(datos_3col[x][2]) for x in range(0,len(datos_3col))]
```

Ahora calculamos la gráfica en *3d* de la función de densidad conjunta a partir de los datos del archivo `xyp.csv`

```python
fig = plt.figure() #Creo la figura
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_trisurf(x, y, z, cmap=cm.jet, linewidth=0)


ax.set_xlim(5,15)
ax.set_ylim(5,25)
ax.set_zlim(0,0.016)

#Nombro los ejes y el grafico
plt.title("Funcion de densidad conjunta a partir de los datos")
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")
fig.tight_layout()
plt.savefig("Graficas/Superficie_densidad_conjunta.png")
```

Ahora calculamos la gráfica en *3d* de la función de densidad conjunta a partir del modelo empleado los parametros calculados anteriormente

```python
def gaussianaR2(x,y,ax,ay,bx,by):
    #tpx=(-(x-ax)**2)/(2*bx)
    #tpy=(-(y-ay)**2)/(2*by)
    return(1/(2*np.pi*bx*by))*np.exp((-(x-ax)**2)/(2*bx)+(-((y-ay)**2)/(2*by)))

#Creo el vector que contiene los valores de la funcion de densidad conjunta del modelo
z2=[]
for i in range(0,len(x)):
    z2.append(gaussianaR2(x[i], y[i], paramX[0], paramY[0], paramX[1], paramY[1]))

#Creo la figura que contiene el grafico
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_trisurf(x, y, z2, cmap=cm.jet, linewidth=0)

ax.set_xlim(5,15)
ax.set_ylim(5,25)
ax.set_zlim(0,0.016)

#Defino los limites de los ejes
ax.set_xlim(5,15.5)
ax.set_ylim(5,25.5)
ax.set_zlim(0,np.max(z2)+0.0005)

#Nombro los ejes
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")
plt.title("Funcion de densidad conjunta a partir del modelo calculado")
fig.tight_layout()
plt.savefig("Graficas/Superficie_densidad_conjunta_del_modelo.png")
```
